```python
#on dataframe
points_df=pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(1,-6)],
        '2nd point':[(5,4),(-6,-3),(7,-4)]
    }
)
points_df
```

```python
def euclid_dist(row):
    pt_A=row['1st point']
    pt_B=row['2nd point']

    return ((pt_A[0]-pt_B[1])**2 + (pt_A[1]-pt_B[0])**2)**0.5
```

```python
points_df['distance']=points_df.apply(euclid_dist,axis=1)
points_df
```
