```python
import numpy as np
import pandas as pd
imdb_df=pd.read_csv(r'C:\Users\HP\Desktop\all folder\pandas\imdb-top-1000.csv')
genres=imdb_df.groupby('Genre')
```

```python
#groupby on multiple cols
duo=imdb_df.groupby(['Director','Star1'])
duo
duo.size()
```

```python
#get_group -- pass tuple
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
#que- find the most earning actor - director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
#find the best(in terms of metascore(avg)) actor- genre combo
imdb_df.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False)
```
