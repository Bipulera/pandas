```python
import numpy as np
import pandas as pd
imdb_df=pd.read_csv(r'C:\Users\HP\Desktop\all folder\pandas\imdb-top-1000.csv')
genres=imdb_df.groupby('Genre')
```

```python
genres.apply(min)
```

```python
#min is a builtin fun .
#apply mostly used for custom function
```

```python
#que- find no.of movies starting with A for each group
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
    
```

```python
genres.apply(foo)
```

```python
#que- find ranking of each movie in the group acoo to imdb ranking
def rank_movie(group):
    group['genre rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank_movie)
```

```python
#que- find normalised rating of imdb group wise
# normalised = ( x-x(min) / x(max)-x(min) )
```

```python
def normalised(group):
    group['normalised_rating']= group['IMDB_Rating']-group['IMDB_Rating'].min() / (group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group

genres.apply(normalised)
```
